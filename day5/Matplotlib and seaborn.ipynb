```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df['Species'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Species', data=df, palette='viridis')
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

image.png

```python
#https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='PetalLengthCm', kde=True, bins=10, color='blue')
plt.title(f'Distribution of PetalLengthCm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='PetalLengthCm', hue='Species', kde=True, bins=20, palette='hls')
plt.title('Histogram of PetalLengthCm by Species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='PetalLengthCm', hue='Species', palette='hls')
plt.title('Histogram of PetalLengthCm by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='PetalLengthCm', color='green')
plt.title(f'Boxplot of PetalLengthCm')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df, palette='hls')
plt.title('PetalLengthCm vs PetalWidthCm')
plt.show()
```

```python
import plotly.express as px

# Create the scatter plot
fig = px.scatter(
    df,
    x='PetalLengthCm',
    y='PetalWidthCm',
    color='Species',
    title='PetalLengthCm vs PetalWidthCm',
    labels={'PetalLengthCm': 'Petal Length (cm)', 'PetalWidthCm': 'Petal Width (cm)'},
    color_discrete_sequence=px.colors.qualitative.Set1  # Custom color palette
)

# Show the plot
fig.show()
```

```python
sns.pairplot(df, hue='Species', diag_kind='kde', palette='hls')
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
