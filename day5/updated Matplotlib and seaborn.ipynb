```python
import numpy as np
from scipy.stats import gaussian_kde

def get_prob_score(val,var1,class_var):
    d = {}
    for cls in df[class_var].unique():
        data = df[df[class_var] == cls][var1]
        kde = gaussian_kde(data)
        pdf_value = kde(val)
        d[cls] = round(pdf_value[0],2)
    return d
```

```python
val = 4.5
var1 = 'PetalLengthCm'
class_var = 'Species'
prob_score = get_prob_score(val,var1,class_var)
prob_score
```

# Titanic

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='Age', hue='Survived', palette='hls')
plt.title('Histogram of Age by Survived')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a KDE plot with relative percentages
plt.figure(figsize=(8, 6))
sns.kdeplot(
    data=df,
    x='Age',
    hue='Survived',
    palette='hls',
    common_norm=False,  # Ensures densities are normalized independently per hue
    fill=True,          # Optional: Add shading under the KDE curve
    alpha=0.5           # Adjust transparency for better visibility
)

# Add title and labels
plt.title('Percentage Histogram of Age by Survival Status', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density (Relative Proportion)', fontsize=12)

# Show the plot
plt.show()
```

```python
# Create a bar plot to visualize the survival count by gender
sns.countplot(data=df, x="gender", hue="Survived", palette="Set2")

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Survival by Gender")

# Show the plot
plt.show()
```

```python
df['gender'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your Titanic dataset
# Calculate proportions
relative_df = (
    df.groupby("gender")["Survived"]
    .value_counts(normalize=True)
    .rename("Proportion")
    .reset_index()
)

# Create the bar plot
sns.barplot(data=relative_df, x="gender", y="Proportion", hue="Survived", palette="Set2")

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Proportion")
plt.title("Proportion of Survival by Gender")

# Show the plot
plt.show()
```

```python
sns.boxplot(data=df, x="gender", y="Age", hue="Survived", palette="Set2")

# Add labels and title
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Age Distribution by Gender and Survival")

# Show the plot
plt.show()
```
